1. Generate a numpy array containing the first 50 natural numbers. Compute another array that is equivalent to the element-wise square root of the first array.
You should do this in two different ways:
(i) Using the math.sqrt() function inside a list comprehension computing explicitly for each element through a loop
(ii) Using the numpy ufunc np.sqrt().
Use the python timeit function to compute and print the time taken for performing (i) and (ii). Note down the speedup/slowdown achieved in the case 

```python
import numpy as np
import math
import timeit

# 1. Generate array of first 50 natural numbers
arr = np.arange(1, 51)

# 2(i). Using math.sqrt() inside list comprehension
def sqrt_with_math():
    return np.array([math.sqrt(x) for x in arr])

# 2(ii). Using numpy's ufunc np.sqrt()
def sqrt_with_numpy():
    return np.sqrt(arr)

# 3. Time comparison using timeit
time_math = timeit.timeit(sqrt_with_math, number=100000)
time_numpy = timeit.timeit(sqrt_with_numpy, number=100000)

# Print results
print("Time using math.sqrt():", time_math)
print("Time using numpy.sqrt():", time_numpy)

# 4. Speedup calculation
speedup = time_math / time_numpy
print("Speedup (math.sqrt / numpy.sqrt):", speedup)
```

4. Create an array big with the first 64 multiples of 12.
(i) Resize the array into a 3 dimensional array tensor of order 8 x 4 x 2

(ii) From tensor select the elements that are divisible by 3 and save them in small_tensor

(iii) Find the difference between the product of the elements of small_tensor and big

```python
import numpy as np

# (i) First 64 multiples of 12
big = np.arange(12, 12*65, 12)

# Resize into 3D tensor 8x4x2
tensor = big.reshape(8, 4, 2)

# (ii) Select elements divisible by 3
small_tensor = tensor[tensor % 3 == 0]

# (iii) Difference between products
prod_big = np.prod(big, dtype=np.int64)
prod_small = np.prod(small_tensor, dtype=np.int64)
difference = prod_small - prod_big

# Print results
print("Big array shape:", big.shape)
print("Tensor shape:", tensor.shape)
print("Small tensor shape:", small_tensor.shape)
print("Product of big:", prod_big)
print("Product of small_tensor:", prod_small)
print("Difference:", difference)
```

5. The parametric function defining a limaçon is:
$$r = r_0 + cos \;θ$$   $$x = r \; cos \;θ$$  $$y = r \; sin \;θ$$ 

Plot the shape of the limaçon for $r_0 = 0.5$, $r_0 = 1.0$, and $r_0 = 1.5$.
Ensure that you plot enough points that the curve is closed and appears smooth.Use a legend to identify which curve is which.

```python
import numpy as np
import matplotlib.pyplot as plt

# Theta range (enough points for smooth curve)
theta = np.linspace(0, 2*np.pi, 1000)

# r0 values
r0_values = [0.5, 1.0, 1.5]

# Plot each curve
plt.figure(figsize=(6,6))

for r0 in r0_values:
    r = r0 + np.cos(theta)
    x = r * np.cos(theta)
    y = r * np.sin(theta)
    plt.plot(x, y, label=f"r0 = {r0}")

# Formatting
plt.title("Limaçon Curves for Different r0")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.axis("equal")   # Equal scaling for x and y
plt.grid(True)
plt.show()
```

6. Create a random numpy integer array of dimension 5 x 10 x 20.
Compute the mean of this array along each of the different axes. Sum the rows of the computed mean. Consider this result as a sequence and :
(i) Plot the 3 results in 3 different subplots stacked horizontally
(ii) Reshape the original array to an one-dimensional array and plot the values of the difference between the mean of this new array and the elements in sequential order. This subplot should be below the subplots in (i) and as wide as the first row (with all the three subplots)


```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Random integer array (5 x 10 x 20)
arr = np.random.randint(1, 100, size=(5, 10, 20))

# Step 2: Compute mean along each axis
mean_axis0 = np.mean(arr, axis=0)  # shape (10,20)
mean_axis1 = np.mean(arr, axis=1)  # shape (5,20)
mean_axis2 = np.mean(arr, axis=2)  # shape (5,10)

# Step 3: Sum rows of the computed mean
sum_axis0 = np.sum(mean_axis0, axis=1)  # shape (10,)
sum_axis1 = np.sum(mean_axis1, axis=1)  # shape (5,)
sum_axis2 = np.sum(mean_axis2, axis=1)  # shape (5,)

# Step 4: Reshape original array into 1D and compute difference
arr_flat = arr.reshape(-1)
mean_flat = np.mean(arr_flat)
diff_seq = arr_flat - mean_flat  # difference sequence

# Step 5: Plotting
fig, axs = plt.subplots(2, 3, figsize=(15, 8), 
                        gridspec_kw={"height_ratios": [1, 1], "width_ratios": [1, 1, 1]})

# Top row → 3 subplots
axs[0, 0].plot(sum_axis0, marker='o')
axs[0, 0].set_title("Sum of rows (mean over axis=0)")

axs[0, 1].plot(sum_axis1, marker='s')
axs[0, 1].set_title("Sum of rows (mean over axis=1)")

axs[0, 2].plot(sum_axis2, marker='^')
axs[0, 2].set_title("Sum of rows (mean over axis=2)")

# Bottom row → wide subplot across all columns
fig.add_subplot(212)  # overwrite bottom row to span all columns
plt.plot(diff_seq, linewidth=0.7)
plt.title("Difference sequence (flattened array - mean)")
plt.xlabel("Index")
plt.ylabel("Difference")

plt.tight_layout()
plt.show()
```
