### 1. Try to load the file in the path `sample_data/california_housing_train.csv` into a `DataFrame` and do basic data exploration to understand the key details about the data.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('california_housing_train.csv')
```

```python
print("Shape of the dataset:", df.shape)
```

```python
print("Column info:")
print(df.info())
```

```python
print("First 5 rows:")
print(df.head())
```

```python
print("Summary statistics:")
print(df.describe())
```

```python
print("Missing values:")
print(df.isnull().sum())
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```
